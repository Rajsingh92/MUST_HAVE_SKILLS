```python
l = [(2,4),(6,8),(10,12)]
for tup in l:
    print(tup)
```

```python
d = {'k1':1,'k2':2,'k3':3}
for item in d:
    print(item)
```

```python
for k,v in d.items():
    print(k)
    print(v) 
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
    if x ==3:
        print('Breaking because x==3')
        break
    else:
        print('continuing...')
        continue
```
